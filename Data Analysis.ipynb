```python
# Fill missing values (NaN) in 'Review Rating' based on the median rating of each product category
#
# ðŸ§© Example:
# Imagine this small dataset:
#   Category    Review Rating
#   Shoes       4.5
#   Shoes       NaN
#   Shoes       5.0
#   Bags        3.0
#   Bags        NaN
#   Bags        4.0
#
# ðŸ‘‰ Step 1: groupby('Category')
# This splits the data into separate groups based on Category:
#   Shoes â†’ [4.5, NaN, 5.0]
#   Bags  â†’ [3.0, NaN, 4.0]
#
# ðŸ‘‰ Step 2: ['Review Rating'] selects only that column from each group.
#
# ðŸ‘‰ Step 3: transform(...) applies a function to each group and 
# returns a column with the same size as the original data.
#
# ðŸ‘‰ Step 4: lambda x: x.fillna(x.median())
# The word 'lambda' is used to define a short, one-line function (an *anonymous function*).
# Here:
#   - 'x' represents one group of 'Review Rating' values (for a single category)
#   - 'x.median()' calculates the median of that group
#   - 'x.fillna(x.median())' replaces any NaN values in that group with the median
#
# So for each category:
#   Shoes â†’ NaN becomes 4.75 (median of 4.5 and 5.0)
#   Bags  â†’ NaN becomes 3.5  (median of 3.0 and 4.0)
#
# âœ… Final result:
#   Category    Review Rating
#   Shoes       4.5
#   Shoes       4.75
#   Shoes       5.0
#   Bags        3.0
#   Bags        3.5
#   Bags        4.0

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform( lambda x: x.fillna(x.median()))
```

```python
# sneak casing methode (it is also a good practice to keep your column names to the lowercase and space with the underscore)
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
# create a new column age_group
import pandas as pd
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
# create new column purchase frequency_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly' : 30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Even 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
